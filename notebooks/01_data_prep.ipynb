# Data Preparation: External Datasets

In this notebook, we prepare the external datasets required for the Agent-as-Shield Recommender System.
- **Clickbait Challenge**: For training the Clickbait Classifier.
- **GoEmotions**: For the Emotional Manipulation Scorer.
- **FakeNewsNet**: For source credibility heuristics.

```python
import os
import pandas as pd
from datasets import load_dataset

processed_dir = "../data/processed"
os.makedirs(processed_dir, exist_ok=True)
print("Ready to download and process datasets.")
```

## 1. GoEmotions Dataset
This dataset contains 58K Reddit comments labeled with 27 emotions. We will use it to train/test our emotional manipulation scorer.

```python
# Load GoEmotions from HuggingFace
print("Loading GoEmotions dataset...")
dataset = load_dataset("go_emotions")

# Convert to pandas dataframe
df_train = pd.DataFrame(dataset['train'])
df_val = pd.DataFrame(dataset['validation'])
df_test = pd.DataFrame(dataset['test'])

# Save processed versions
df_train.to_csv(os.path.join(processed_dir, "goemotions_train.csv"), index=False)
df_test.to_csv(os.path.join(processed_dir, "goemotions_test.csv"), index=False)
print("GoEmotions saved to data/processed/")
```

## 2. Clickbait Challenge & FakeNewsNet
*(These datasets require manual download from their respective sources: https://www.clickbait-challenge.org/ and https://github.com/KaiDMML/FakeNewsNet)*

Once downloaded to `data/raw/`, we can load and preprocess them here.

```python
def process_clickbait():
    raw_path = "../data/raw/clickbait_data.csv"
    if os.path.exists(raw_path):
        df = pd.read_csv(raw_path)
        # Perform preprocessing steps here...
        df.to_csv(os.path.join(processed_dir, "clickbait_processed.csv"), index=False)
        print("Clickbait dataset processed.")
    else:
        print(f"Raw dataset not found at {raw_path}. Please download it manually.")

process_clickbait()
```
